<b> Merging the train and the test test dataset and saving it in a dict<b>

```python
listOfFiles = os.listdir('UCRArchive_2018/')
pattern = "*.tsv"
dict_conc_list_sets ={}
for entry in listOfFiles:
    if '.' in entry :
        
       continue
    else :    
        #print("this is the dir{}".format(entry))
        for file in os.listdir('UCRArchive_2018/'+ entry):
            if (len(file.split('_'))==1):
                continue
            else:
                file_name =file.split('_')[0]
                x = file.split('_')[1] 
            #print("this is the file name{}",file)
            #if not (fnmatch.fnmatch(file,pattern)):
                #continue
            #print(x)  
            if 'TEST' in x :
                    #print('This is the file{}'.format(file))
                    test = pd.read_csv('UCRArchive_2018/'+entry+"/"+file, sep='\t', header = None)
                    
            if 'TRAIN' in x:
                    #print ('This is the file{}'.format(file))
                    train = pd.read_csv('UCRArchive_2018/'+entry+"/"+file, sep='\t',header = None)
           
    dict_conc_list_sets[entry]= pd.concat([train, test], axis=0)
#print (dict_conc_list_sets.keys)        
            
           
       
  


```

<b> After adding all the files we divide the data into 3 dictonires training, testing and validation<b> 

```python
for key in dict_conc_list_sets:
    data = dict_conc_list_sets[key]
    y = data.iloc[:,[0]]
    #print(y.shape)
    x = data.drop([0],axis = 1)
    #print(x.shape)
    x_mean = x.mean()
    x_std = x.std()
    x_standardized = (x - x_mean) / x_std
    data = pd.concat([y,x_standardized ], axis=1)
    dict_conc_list_sets[key] = data
    #print (data.shape)
    #data.head()



```

```python
train_dic ={}
test_dic = {}
validation_dic ={}
for key in dict_conc_list_sets:
    data = dict_conc_list_sets[key]
    #print(data.head())
    #print("this is the initial shape of the data :",data.shape)
    idx = data.groupby(0).apply(lambda x: x.sample(frac=0.7, random_state = 3116)).index.get_level_values(1)
    train = data.iloc[idx, :].reset_index(drop = True)
    test = data.drop(idx).reset_index(drop = True)
    d2 = test 
    id_2 = d2.groupby(0).apply(lambda x: x.sample(frac=0.5, random_state = 3116)).index.get_level_values(1)
    test = d2.iloc[id_2, :].reset_index(drop = True)
    validation = d2.drop(id_2).reset_index(drop = True)
    
    # after splitting add every shape in its right place in the dict
    train_dic[key] = train
    test_dic[key] = test
    validation_dic[key] = validation 
    
   




   
```
